```python
import numpy as np
import matplotlib.pyplot as plt

print(" Aufgabe 1")


f = lambda t, s: 0.5*np.exp(-t) 
h = 0.5 
t = np.arange(0, 7 + h, h) 
s0 = 0 

s = 0.5*np.exp(-t)
s[0] = s0

for i in range(0, 7):
    s[i] = 0.5 * np.exp(-i)

print(s)
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--')
plt.show()
```
